```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Dec 13 19:52:23 2022

@author: eli
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy import fft

ax = []
ay = []
az = []
dt = []

#f = open('/home/eli/Documents/Shaker_project_V2/teensy_data_aquisition/Shaker_v2_python_plotter/test_data_3.csv','r')
f = open('/home/eli/Documents/Shaker_project_V2/shaker_v2/shaker_rig_code/data/freq_70_vol_20.csv','r')
for line in f:
    line_list = line.strip('[]\n ').split(',',3)
    ax.append(float(line_list[0].strip('.')))    
    ay.append(float(line_list[1].strip('.')))
    az.append(float(line_list[2].strip('.')))
    dt.append(float(line_list[3].strip('.')))

f.close()
```

```python
sample_period_us = np.mean(dt)
print (sample_period_us)
sample_rate = 1 / (sample_period_us * 10**-6)
#print( sample_frequency)
num_samples = len(dt)
print(dt[100])
```

```python
dft = fft.rfft(az[:10000]) # scipy fft library (import scipy.fft^)
dft_freq = fft.rfftfreq(10000, 1/sample_rate)
```

```python
plt.plot(dft_freq,abs(dft))
plt.xlim([0,3000])
plt.yscale('log')
plt.rcParams['figure.dpi'] = 400
plt.show()
```

```python

plt.plot(ax[1000:])
plt.plot(ay[1000:])
plt.plot(az[1000:])
plt.plot(dt[1000:])
plt.xlim([1000,2000])
plt.ylim([-500,500])
plt.show()
```

```python
len_frame = 10000
num_frames = 10
data_frames = []
dft_frames = []
abs_dft_frames = []
for i in range(num_frames):
    data_frames.append(az[i*len_frame:(i+1)*len_frame])
    dft_frames.append(fft.fft(data_frames[i]))
    abs_dft_frames.append( list(map(abs,dft_frames[i])))
dft_frames_freq = fft.fftfreq(len_frame, 1/sample_rate)
```

```python
print(len(data_frames[0]))
#plt.plot(dft_frames_freq, abs_dft_frames[3])
print(len(dft_frames_freq))
for i in range(num_frames):
    plt.plot(dft_frames_freq,abs_dft_frames[i])

plt.xlim([0,1000])
plt.yscale('log')
plt.rcParams['figure.dpi'] = 400
```
